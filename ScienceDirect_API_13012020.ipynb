```python
import requests
import json
import csv
import datetime
import sys
import pandas as pd

URL = 'https://api.elsevier.com/content/search/sciencedirect/'
```

```python
#query params
in_httpAccept ='application/json' #can choose type : json, atom+xml, xml
in_apiKey = 'feab67568ad17f151ebc398b63ecf22d'
in_date = '2010-2019'
in_count = 100
in_sort = 'coverDate' #options: coverDate, relevance
```

```python
#to make query
#keyword = ''
#stopword = 

in_query = "penguin" #https://dev.elsevier.com/tips/ScienceDirectSearchTips.htm
```

```python
PARAMS = {'httpAccept' : in_httpAccept, 
          'apiKey' : in_apiKey,
          'query' : in_query,
          'date':in_date, 
          'count' : in_count,
          'sort':in_sort}
```

```python
# sending get request and saving the response as response object 
r = requests.get(url = URL, params = PARAMS) 
print(r)
```

```python
# extracting data in json format 
data = r.json()
```
